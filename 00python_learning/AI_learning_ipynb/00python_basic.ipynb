```python
plt.plot([1,2,3,4], [1,4,9,16], 'r^')
plt.title("my first plot")
plt.xlabel('x')
plt.ylabel('y')
plt.text(2, 4.5, 'Second')
plt.text(1, 10, r'$y=x^2$', fontsize=20, bbox={'facecolor':'yellow', 'alpha':0.2})
plt.grid(True)
plt.legend(["First Series"], loc=4)
plt.show()
```

```python
x = np.linspace(1, 10, 50)
y1 = 2*x + 1
y2 = x**2
plt.figure()
plt.plot(x, y1)
plt.show()

plt.figure(num = 1, figsize = (8, 5))
plt.plot(x, y1)
plt.plot(x, y2, 'r--', linewidth = 3.0)
plt.show()

plt.subplot(211)
plt.plot(x, y1)
plt.subplot(212)
plt.plot(x, y2)
plt.show()
```

```python
import cv2
```

```python
img = cv2.imread('cat.jpg', cv2.IMREAD_COLOR)
resize = cv2.resize(img, (400,400), interpolation=cv2.INTER_CUBIC)
cv2.imwrite('re.jpg', resize)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imwrite('re.jpg', gray)
canny = cv2.Canny(gray,50,150)
cv2.imwrite('re.jpg',canny)
gray_lap = cv2.Laplacian(img, cv2.CV_16S, ksize=3)
cv2.imwrite('re.jpg',gray_lap)
img_sobel_x = cv2.Sobel(gray,cv2.CV_64F,1,0,ksize=3)
img_sobel_y = cv2.Sobel(gray,cv2.CV_64F,0,1,ksize=3)
cv2.imwrite('re.jpg',img_sobel_x)
cv2.imwrite('re.jpg',img_sobel_y)
```
